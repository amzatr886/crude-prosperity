### Oil Producing Countries

Import libraries

```python
# Load pandas for data exploration
import pandas as pd
# load seaborn for plotting and styling
import seaborn as sns
# load matplotlib for plotting
import matplotlib.pyplot as plt
```

Data loading

```python
# Load the data
oil_producer = pd.read_csv("../data/final/oil_producing_countries.csv")
# preview the data
display(oil_producer)
```

```python
# world bank data
gdp = pd.read_csv("../data/processed/gdp_clean_final.csv")
display(gdp)
```

Data Analysis

```python
"""
 This code Calculate the mean of the GDP over the years.
"""
# select the columns with the years and convert it to string
numerical = [str(x) for x in range(2002, 2022)]
# calculate the mean of the gdp
gdp_mean = gdp[numerical].mean().reset_index()
# rename the columns
gdp_mean.columns = ["year", "avg-gdp"]
# convert the avg-gdp column value to tillions
gdp_mean["avg-gdp"] = gdp_mean["avg-gdp"] / 1000000000

# calculate the growth rate
gdp_mean["growth-rate"] = gdp_mean["avg-gdp"].pct_change()
display(gdp_mean)
```

```python
"""
Plot a line graph of the average GDP of oil producing countries over the years.

Parameters:
- data: DataFrame containing the average GDP data
- x: Column name for the x-axis (Year)
- y: Column name for the y-axis (Avg-gdp)
- marker: Marker style for the data points
Returns:
None
"""
sns.lineplot(data=gdp_mean, x="year", y="avg-gdp", marker="o")
plt.title("Average GDP \n Current US($) from 2002 to 2021")
plt.xticks(rotation=45)
plt.xlabel("YEAR")
plt.ylabel("GDP \n IN BILLIONS ")

_yticks = gdp_mean["avg-gdp"].to_list()
_xticks = plt.xticks()[0]

# display(plt.xticks()[0])
# display(plt.yticks())

recession = plt.annotate("2008 reccession impact",
                    xy=(_xticks[7],_yticks[8]), 
                    xytext=(_xticks[3], _yticks[10]),
                    arrowprops=dict(facecolor="green", shrink =0.05)
)                    

oil = plt.annotate("global supply glut of oil ",
                    xy=(_xticks[13],_yticks[13]), 
                    xytext=(_xticks[10], _yticks[19]),
                    arrowprops=dict(facecolor="red", shrink =0.05)                    
)
covid = plt.annotate("covid 19 pandemic impact ",
                    xy=(_xticks[18],_yticks[17]), 
                    xytext=(_xticks[18], _yticks[19]),
                    arrowprops=dict(facecolor="orange", shrink =0.05)
)
plt.show()
```

```python
"""
    Adds a new column to the oil_producer DataFrame to indicate whether a country is from Africa or from other continents.

    Parameters:
    oil_producer (DataFrame): The DataFrame containing information about oil producers.

    Returns:
    DataFrame: The updated DataFrame with the 'continent' column added.
"""
    
# create a list for african countries
africa = [
    "Nigeria",
    "Libya",
    "Angola",
    "Algeria",
    "Egpyt",
    "Republic of congo",
    "Gabon",
    "Sudan",
    "Chad",
    "Cameroon",
    "Equatorial guinea",
    "Tunisia",
    "Ivory cost",
    "Democratic republic of Congo",
]

# create a new column to indicate countries from africa continent and others continents
oil_producer["continent"] = oil_producer["country"].apply(
    lambda x: "africa" if x in africa else "others"
)
# preview the data
display(oil_producer)
```

```python
# filter by countrires
gdp[gdp["code"]== "USA"]
```

Data Transformation

```python
"""
This code calculates the mean GDP of each oil producing country for the years 2002 to 2021.
It then calculates the percentage change in GDP and displays the results.
"""

# convert the year column to string
numeric_columns = [str(x) for x in range(2002, 2022)]
# calculate the mean of the gdp of each country for 2002 to 2021
mean_gdp = oil_producer[numeric_columns].mean().reset_index()

# rename the columns
mean_gdp.columns = ["year", "avg-gdp"]
#convert mean_gdp to billions
mean_gdp["avg-gdp"] = mean_gdp["avg-gdp"]/ 1000000000
# calculate the percentage change in gdp
mean_gdp["growth-rate"] = mean_gdp["avg-gdp"].pct_change()
# preview data
display(mean_gdp)
```

```python
"""
Calculate the mean GDP of each country for the years 2002 to 2021, grouped by continent.

Parameters:
- oil_producer: DataFrame containing the data of oil producers
- numeric_columns: List of column names representing the years 2002 to 2021

Returns:
- mean_gdp_continent: DataFrame containing the mean GDP of each country for the years 2002 to 2021, grouped by continent
"""

# convert year column to string
numeric_columns = [str(x) for x in range(2002, 2022)]
# calculate the mean of the gdp of each country for 2002 to 2021 group by continent
mean_gdp_continent = oil_producer.groupby("continent")[numeric_columns].mean().reset_index()

# preview data
display(mean_gdp_continent)
```

```python
"""
This code transposes the data in `mean_gdp_continent`, renames columns, converts data types, and calculates growth rates.
"""

# transpose the data
africa_vs_other = mean_gdp_continent.transpose().reset_index()
# select column index 0
africa_vs_other.columns = africa_vs_other.iloc[0]
# drop column index 0
africa_vs_other.drop(0, inplace=True)
# rename continent column to year
africa_vs_other.rename(columns={"continent": "year"}, inplace=True)
# convert year column to integer
africa_vs_other["year"] = africa_vs_other["year"].astype(int)
# transform gdp value to billions
africa_vs_other["africa"] = africa_vs_other["africa"] / 1000000000
africa_vs_other["others"] = africa_vs_other["others"] / 1000000000
# calculate the percentage change in africa gdp
africa_vs_other["africa_growth_rate"] = africa_vs_other["africa"].pct_change()
# calculate the percentage change in other gdp
africa_vs_other["other_growth_rate"] = africa_vs_other["others"].pct_change()
# preview data
display(africa_vs_other)
```

Data Visualization

```python
"""
Plot a line graph of the average GDP of oil producing countries over the years.

Parameters:
- data: DataFrame containing the average GDP data
- x: Column name for the x-axis (Year)
- y: Column name for the y-axis (Avg-gdp)
- marker: Marker style for the data points
Returns:
None
"""
sns.lineplot(data=mean_gdp, x="year", y="avg-gdp", marker="o")
plt.title("Average GDP (Oil Producing Countries) \n Current US($) from 2002 to 2021")
plt.xticks(rotation=45)
plt.xlabel("YEAR")
plt.ylabel("GDP \n IN BILLIONS ")

_yticks = mean_gdp["avg-gdp"].to_list()
_xticks = plt.xticks()[0]

# display(plt.xticks()[0])
# display(plt.yticks())

rect = plt.annotate("2008 reccession impact",
                    xy=(_xticks[7],_yticks[8]), 
                    xytext=(_xticks[3], _yticks[10]),
                    arrowprops=dict(facecolor="green", shrink =0.05)
)                    

economic = plt.annotate("global supply glut of oil ",
                    xy=(_xticks[13],_yticks[13]), 
                    xytext=(_xticks[10], _yticks[19]),
                    arrowprops=dict(facecolor="red", shrink =0.05)                    
)
covid = plt.annotate("covid 19 pandamic impact ",
                    xy=(_xticks[18],_yticks[17]), 
                    xytext=(_xticks[18], _yticks[19]),
                    arrowprops=dict(facecolor="orange", shrink =0.05)
)
plt.show()
```

```python
"""
 Plots the growth rate trend of Oil Producing Countries
"""
sns.lineplot(data=mean_gdp, x="year", y="growth-rate", marker="o", color="orange")
plt.title("GDP Growth Rate Trends in Oil-Producing Countries\n (2003-2021)")
plt.xticks(rotation=45)
plt.xlabel("YEAR")
plt.ylabel("GDP GROWTH RATE")
plt.show()
# save the plot
plt.savefig("../documentation/figure/gdp_growth_rate.png")
```

```python
"""
Plot a line graph comparing the average GDP of Africa and other oil Producing countries continents over the years.

Parameters:
- data: DataFrame containing the data for the line graph.
- x: Column name for the x-axis values.
- y: Column name for the y-axis values.
- marker: Marker style for the data points on the line.
- label: Label for the line in the legend.
- color: Color of the line.
- legend: Boolean indicating whether to display the legend.
"""
sns.lineplot(
    data=africa_vs_other,
    x="year",
    y="africa",
    marker="o",
    label="Africa",
    color="orange",
    legend=True,
)
sns.lineplot(
    data=africa_vs_other,
    x="year",
    y="others",
    marker="o",
    color="green",
    label="Others",
    legend=True,
)
plt.title("Africa vs Others continent \n (oil producing countries) GDP")
plt.xticks(africa_vs_other["year"].unique().tolist(), rotation=45)
plt.xlabel("YEAR")
plt.ylabel("AVG GDP IN BILLIONS")
plt.show()
```

```python
"""
Plot a line graph comparing the growth rate of Africa and other continents over the years.

Parameters:
- data: DataFrame containing the data for the line graph.
- x: Column name for the x-axis values.
- y: Column name for the y-axis values.
- marker: Marker style for the data points on the line.
- label: Label for the line in the legend.
- color: Color of the line.
- legend: Boolean indicating whether to display the legend.
"""
sns.lineplot(
    data=africa_vs_other,
    x="year",
    y="africa_growth_rate",
    marker="o",
    label="Africa",
    color="orange",
    legend=True,
)
sns.lineplot(
    data=africa_vs_other,
    x="year",
    y="other_growth_rate",
    marker="o",
    color="green",
    label="Others",
    legend=True,
)
plt.title("Africa vs Others continent growth rate \n (oil Producing countries)")
plt.xticks(africa_vs_other["year"].unique().tolist(), rotation=45)
plt.xlabel("YEAR")
plt.ylabel("GROWTH RATE")
plt.show()
```
