## insight-01: Clean World  GDP Data: <br>
drop columns from 1960 t0 1999 <br>
Rename columns <br>
export data to processed folder <br>

```python
import pandas as pd  # data cleaning and processing
import csv  # for data exporting and qouting character
```

```python
gdp = pd.read_csv(
    "../data/raw/world-gdp.csv"
)  # create a data frame named  GDP for the world gdp data for eazy access and cleaning process
```

```python
gdp.head()
```

```python
gdp_renamed = gdp.rename(
    columns={
        "Country Name": "country",
        "Country Code": "code",
        "Indicator Name": "currency",
        "Indicator Code": "indicator_code",
    }
)  # rename the columns to make them more readable
```

```python
gdp_renamed.head()
```

```python
# drop years before 2000
gdp_v1 = gdp_renamed.drop(gdp_renamed.columns[4:44], axis=1)  # drop all  years columns before 2000
```

```python
gdp_v1.head()
```

```python
gdp_v2 = gdp_v1.drop(
    columns=["Unnamed: 67", "indicator_code"]
)  # drop the columns that are not needed for the analysis
```

```python
gdp_v2.head()
```

```python
gdp_clean_final = gdp_v2.to_csv(
    "../data/processed/gdp_clean_final.csv", index=None, quoting=csv.QUOTE_ALL, quotechar='"'
)  # save the cleaned data to a csv file for processing in the next step
```
